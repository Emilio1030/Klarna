  9. status_Nth_archived_M_Nm: `status of the last Nth archived invoice between M and N months ago`
    `0 through 5, with 1 being the most commom status... perhaps:`

      -  `1 means paid fully in short term installments`

      -  `2 means paid fully in longer term installments`

      -  `3 means paid fully in installments with fee due to 
        delays in payment`

      -  `4 means not paid in full (is not present in the train dataset)`
        
      -  `5 means not paid at all`

```python
# has_paid feature
has_paid_true = len(data['has_paid'][data['has_paid'] == True]) / len(data)
has_paid_missing = data['has_paid'].isna().sum()
unique_values = data['has_paid'].unique()

print(f'we have either True or False {unique_values }')
print(f'{round(has_paid_true*100, 2)}% are True values and we have {has_paid_missing} missing values')
```

```python
# recovery_debt
rec_debt = data['recovery_debt'].isna().sum()
client_defaults_amounts = len(data['recovery_debt'][data['recovery_debt'] > 0])
print(f'we have {rec_debt} missing values')
print(f'we have {client_defaults_amounts} with default amounts')
```

```python
# time_hours
print(data['time_hours'].describe())
time_hours_missing = data['time_hours'].isna().sum()
print(f'we have {time_hours_missing} missing values')
```

```python
# worst_status_active_inv'
worst_status_active_inv_missing = len(data['worst_status_active_inv'].isna()) / len(data)
print(f'{worst_status_active_inv_missing*100}% of the values are missing')
```
